### Matrix for abstracts

```python
# Functions for matrix creation

def clear_text(text, lowerize=True):

    pat = re.compile(r'[^A-Za-z0-9 \-\n\r.,;!?А-Яа-я]+')
    cleared_text = re.sub(pat, ' ', text)

    if lowerize:
        cleared_text = cleared_text.lower()

    tokens = cleared_text.split()
    return tokens


def make_substrings(tokens, k=4):

    for i in range(max(len(tokens) - k + 1, 1)):
        yield ' '.join(tokens[i:i + k]


def save_matrix(matrix, name="data/relevance_matrix.txt" ):
    np.savetxt(name, matrix)
```

```python
# Functions that creates matrix for 1st matrix
def get_relevance_matrix(texts, strings):

    matrix = np.empty((0, len(strings)), float)
    prepared_text_tokens = [clear_text(t) for t in texts]

    prepared_string_tokens = [clear_text(s) for s in strings]
    prepared_strings = [' '.join(t) for t in prepared_string_tokens]
    
    i = 0
    for text_tokens in prepared_text_tokens:
        print('\r',i, len(prepared_text_tokens), end = '')
        i += 1
        ast = base.AST.get_ast(list(make_substrings(text_tokens)))
        row = np.array([ast.score(s) for s in prepared_strings])
        matrix = np.append(matrix, [row], axis=0)

    return matrix
```

```python
# Calculating final matrix
matrix = []
i = 0
for k_tokens in prepared_keywords:
    res_row = np.zeros(len(prepared_strings))
    print('\r',i, len(prepared_keywords), end = '')
    i += 1
    for token in k_tokens:
        res_row = np.maximum(res_row, pre_res.get(' '.join(token), np.zeros(len(prepared_strings))))
    matrix.append(res_row)
```

## Matrix from abstracts and keywords with different agregation functions

```python
def aggr_func_base(i, j, matrix, s_matrix):
    pre_c = np.sum(s_matrix, axis=0) + 1
    return np.sum(matrix[i,:] * matrix[j,:] / pre_c)

def aggr_func_mine(i, j, matrix, s_matrix):
    pre_c = np.sum(s_matrix, axis=1) + 1 
    return np.sum(matrix[i,:] * matrix[j,:] / min(pre_c[i], pre_c[j]))

def matrix_creation(matrix, alpha=0.38, func=aggr_func_base):
    n = matrix.shape[0]
    res = np.zeros((n, n))
    s_matrix = matrix > alpha
    for i in range(n):
        for j in range(i, n):
            l_res = func(i, j, matrix, s_matrix)
            print('\r', i, j, n, l_res,  end = '')
            res[i][j] = l_res
            res[j][i] = l_res
    return res
```

```python
relevance_matrix = np.loadtxt("data/relevance_matrix1.txt").T
types_matrix = matrix_creation(relevance_matrix, alpha=0.2)
    
np.savetxt('data/NCrossMatrix1_1.txt', types_matrix)
```
